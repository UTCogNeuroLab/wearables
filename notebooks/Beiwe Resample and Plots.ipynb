# Beiwe Fitbit Data

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
homeDir = '/Users/megmcmahon/'
fitbitDataDir = homeDir + 'Box/CogNeuroLab/Wearables/data/fitbit/beiwe/'
fitbitDataFiles = os.listdir(fitbitDataDir)
```

```python
fitbitDataFiles
```

```python
df = pd.read_csv(fitbitDataDir + fitbitDataFiles[0], parse_dates = True, index_col = 'date')
df
```

```python
fig, ax = plt.subplots(figsize=(12, 12))

# Add x-axis and y-axis
ax.scatter(df.index,
       df['activities_steps'],
       color='purple')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Steps",
       title="Fitbit Activities")

plt.show()
```

```python
for file in fitbitDataFiles:
    df = pd.read_csv(fitbitDataDir + file, parse_dates = True, index_col = 'date')
    print(df)
    dfSteps = df[['activities_steps']]
    print(dfSteps)
    dfHeart = df[['activities_heart']]
    print(dfHeart)
    
    dfSteps.resample('1T').sum().to_csv(homeDir + 'Box/CogNeuroLab/Wearables/data/fitbit/' + file.split('.csv')[0] + '_minuteStepsNarrow.csv', header = ['Steps'], index_label = ['ActivityMinute'])
    dfHeart.resample('1T').sum().to_csv(homeDir + 'Box/CogNeuroLab/Wearables/data/fitbit/' + file.split('.csv')[0] + '_heartrate_1min.csv', header = ['Value'], index_label = ['Time'])
```

## Steps

```python
fig, axs = plt.subplots(10,1, figsize=(15, 60), facecolor='w', edgecolor='k')

axs = axs.ravel()
i = 0
for file in fitbitDataFiles[0:10]:
    df = pd.read_csv(fitbitDataDir + file, parse_dates = True, index_col = 'date')

    if len(df.index) > 0:
        axs[i].scatter(df.index, df['activities_steps'])
        axs[i].set_title(file.split('.csv')[0] + ' , ' + str(df.index[-1] - df.index[0]) )
        i = i+1
    else:
        print('missing data: ' + file.split('.csv')[0])
```

```python
fig, axs = plt.subplots(4,1, figsize=(15, 30), facecolor='w', edgecolor='k')

axs = axs.ravel()
i = 0
for file in fitbitDataFiles[11:15]:
    df = pd.read_csv(fitbitDataDir + file, parse_dates = True, index_col = 'date')
    if len(df.index) > 0:
        axs[i].scatter(df.index, df['activities_steps'])
        axs[i].set_title(file.split('.csv')[0] + ' , ' + str(df.index[-1] - df.index[0]) )
        i = i+1
    else:
        print('missing data: ' + file.split('.csv')[0])
```
